```python
from langchain_community.tools import DuckDuckGoSearchRun
from langchain.agents import initialize_agent, AgentType
from langchain_community.llms import Ollama
from langgraph.graph import StateGraph, END
from typing import TypedDict
```

```python
# Inicialize o modelo Gemma 3
llm = Ollama(model="gpt-oss:20b")
```

```python
# Faça uma consulta
resposta = llm.invoke("Setembro chove em Ribeirão Preto?")
print(resposta)
```

```python
search = DuckDuckGoSearchRun()
search.invoke("Obama's first name?")
```

```python
# Função para processar perguntas
def pesquisar_e_responder(state):
    pergunta = state["pergunta"]
    resultado_busca = search.invoke(pergunta)
    print(f"Resultado da busca: {resultado_busca}")
    resposta = llm.invoke(f"Use essa informação para responder a pergunta do usuário: {resultado_busca}.\nLimite-se a responder apenas a pergunta: {pergunta}")
    return {"resposta": resposta}
```

```python
input = {"pergunta": "Qual é a capital do Brasil?"} 
pesquisar_e_responder(input)
```

```python
class PerguntaState(TypedDict):
    pergunta: str
    resposta: str
```

```python
# Defina o grafo de estados
graph = StateGraph(PerguntaState)
graph.add_node("pesquisar_e_responder", pesquisar_e_responder)
graph.add_edge("pesquisar_e_responder", END)
graph.set_entry_point("pesquisar_e_responder")
```

```python
# Compile e execute
agent = graph.compile()
state = {"pergunta": "Me traga informações sobre Mateus Tarcinalli Machado"}
result = agent.invoke(state)
print(result["resposta"])
```
